```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

#### Import dataset, create new one  and print only 20 first positions of the dataset, plot the graph with results

```python
twitter_stars = pd.read_csv('/home/hubert/Exercises/twitter_celebrities.csv')
```

```python
twitter_stars.head(20)
```

```python
new_data = pd.DataFrame(twitter_stars).head(5)
```

```python
new_data
```

#  Plot the graph

```python
labels = new_data['Account Owner']
values = new_data['Followers(in M)']

plt.title("TOP 5 Account Owner",pad=20, fontsize=22)
plt.xlabel("Account Owner", labelpad=16, fontsize=16)
plt.ylabel("Followers(in M)", labelpad=16, fontsize=16)



plt.bar(labels, values, color=['yellow', 'red', 'green', 'blue', 'orange'])

plt.show()
```

#### Create new dataframe with all values needed to sum column value 

```python
all_data = pd.DataFrame(twitter_stars)
```

#### Group dataset by country and number of followers, plot the pie chart

```python
all_data.groupby('Country')['Followers(in M)'].sum()
```

```python
all_data.groupby('Country')['Followers(in M)'].sum().head(10).plot(kind='pie', autopct='%1.1f%%')

plt.title("Percentage of the followers by Country (TOP 10)", fontsize=16)
plt.ylabel("")
plt.xlabel("Country",labelpad=16, fontsize=16)

plt.show()
```

#### Group dataset by occupation (top 10 places) and number of followers, plot the pie chart

```python
all_data.groupby('Occupation')['Followers(in M)'].sum().head(10).plot(kind='pie', autopct='%1.1f%%')

plt.title("Percentage of the followers by profession TOP 10", fontsize=16)
plt.ylabel("")

plt.xlabel("Occupation",labelpad=16, fontsize=16)

plt.show()
```

#### Group the dataset, sum values of 'Followers' column and sort from the highest to the lowest value

```python
all_data.groupby('Country')['Followers(in M)'].sum().sort_values(ascending=False).plot.bar(color=['pink', 'red', 'green', 'blue', 'cyan', 'green', 'silver', 'yellow', 'green','gray'])

plt.title("Followers(in M) by Country",pad=20, fontsize=22)
plt.xlabel("Country", labelpad=16, fontsize=16)
plt.ylabel("Followers(in M)", labelpad=16, fontsize=16)




plt.show()
```
